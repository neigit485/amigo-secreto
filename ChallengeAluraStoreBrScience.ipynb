### Importação dos dados



```python
from IPython.display import display
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

display(loja1.head())
display(loja2.head())
display(loja3.head())
display(loja4.head())
```

#1. Análise do faturamento



```python
# 2. URLs dos CSVs (raw)
base = ("https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/")
url1 = base + "loja_1.csv"
url2 = base + "loja_2.csv"
url3 = base + "loja_3.csv"
url4 = base + "loja_4.csv"

# 3. Carregamento dos dados
loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# 4. Cálculo de faturamento (somatório da coluna 'Preço'; frete desconsiderado)
fat_loja1 = loja1['Preço'].sum()
fat_loja2 = loja2['Preço'].sum()
fat_loja3 = loja3['Preço'].sum()
fat_loja4 = loja4['Preço'].sum()

# 5. Consolidado em um DataFrame para fácil visualização/manipulação posterior
faturamento = pd.DataFrame({
    'Loja':   ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4'],
    'Faturamento (R$)': [fat_loja1, fat_loja2, fat_loja3, fat_loja4]
})

# 6. Exibe o resultado no console
print(faturamento)
```

# 2. Vendas por Categoria

```python
# 1. Junta as quatro lojas em um único DataFrame
todas_as_lojas = pd.concat([loja1, loja2, loja3, loja4], ignore_index=True)

# 2. Cria o resumo por categoria
resumo_por_categoria = (
    todas_as_lojas
      .groupby('Categoria do Produto')               # agrupa por categoria
      .agg(Preço=('Preço', 'first'),                 # preço “base” (primeira ocorrência)
           **{'Quantidade de vendida': ('Preço', 'size')},   # total de vendas (linhas)
           **{'Faturamento das Vendas do Produto': ('Preço', 'sum')})  # soma dos preços
      .reset_index()
)

# 3. Exibe o resultado
resumo_por_categoria
```

```python
# 1. Função auxiliar: devolve faturamento por categoria em um DataFrame
def faturamento_por_categoria(df):
    return (df
            .groupby('Categoria do Produto', as_index=False)['Preço']
            .sum()
            .rename(columns={'Preço': 'Faturamento com o Produto (R$)'}))

# 2. Gera o resumo para cada loja
fat_loja1 = faturamento_por_categoria(loja1)
fat_loja2 = faturamento_por_categoria(loja2)
fat_loja3 = faturamento_por_categoria(loja3)
fat_loja4 = faturamento_por_categoria(loja4)

# 3. (Opcional) Grava tudo em um único arquivo Excel, uma aba por loja
#    Se ainda não tiver, instale:  !pip install openpyxl
with pd.ExcelWriter('faturamento_por_loja.xlsx') as writer:
    fat_loja1.to_excel(writer, sheet_name='Loja1', index=False)
    fat_loja2.to_excel(writer, sheet_name='Loja2', index=False)
    fat_loja3.to_excel(writer, sheet_name='Loja3', index=False)
    fat_loja4.to_excel(writer, sheet_name='Loja4', index=False)

# 4. Visualização rápida no console (por exemplo, Loja1)
print(fat_loja2)
```

```python
todas_as_lojas = (
    pd.concat([                                  # empilha verticalmente …
        loja1.assign(Loja='Loja 1'),             # … marcando qual é a loja
        loja2.assign(Loja='Loja 2'),
        loja3.assign(Loja='Loja 3'),
        loja4.assign(Loja='Loja 4')
    ], ignore_index=True)                        # reindexa linhas
)

# Visualização completa
display(todas_as_lojas)

# (Opcional) apenas as 5 primeiras linhas de cada loja, para não lotar a tela
# display(todas_as_lojas.groupby('Loja').head())
```

```python
# 1. Cria um arquivo Excel com as 4 lojas lado a lado
with pd.ExcelWriter('faturamento_todas_lojas.xlsx', engine='openpyxl') as writer:
    # Escreve cada loja em posições diferentes (colunas)
    fat_loja1.to_excel(writer, sheet_name='Resumo', startcol=0, index=False)
    fat_loja2.to_excel(writer, sheet_name='Resumo', startcol=3, index=False)
    fat_loja3.to_excel(writer, sheet_name='Resumo', startcol=6, index=False)
    fat_loja4.to_excel(writer, sheet_name='Resumo', startcol=9, index=False)

# 2. Visualiza no console (uma embaixo da outra)
print("LOJA 1")
print(fat_loja1)
print("\nLOJA 2")
print(fat_loja2)
print("\nLOJA 3")
print(fat_loja3)
print("\nLOJA 4")
print(fat_loja4)
```

# 3. Média de Avaliação das Lojas


```python
# 1. Calcula a média de avaliação de cada loja
media_avaliacao = {
    'Loja 1': loja1['Avaliação da compra'].mean(),
    'Loja 2': loja2['Avaliação da compra'].mean(),
    'Loja 3': loja3['Avaliação da compra'].mean(),
    'Loja 4': loja4['Avaliação da compra'].mean()
}

# 2. Converte em DataFrame para visualização amigável
media_avaliacao_df = (
    pd.Series(media_avaliacao)
      .reset_index()
      .rename(columns={'index': 'Loja', 0: 'Média de Avaliação'})
      .round(2)                      # deixa com 2 casas decimais
)

# 3. Exibe o resultado
print(media_avaliacao_df)
```

# 4. Produtos Mais e Menos Vendidos

```python
# 1. Conta a ocorrência de cada produto
mais_vendidos = (
    todas_as_lojas['Produto']
      .value_counts()
      .reset_index()
)

# 2. Renomeia as colunas para clareza
mais_vendidos.columns = ['Produto', 'Quantidade Vendida']

# 3. Exibe a lista dos mais vendidos
print("--- Produtos Mais Vendidos (Geral) ---")
print(mais_vendidos)

# 4. (Opcional) Salva em um arquivo Excel
mais_vendidos.to_excel("produtos_mais_vendidos.xlsx", index=False)
```

```python
# Menos vendidas
menos_vendidos = mais_vendidos.sort_values(by='Quantidade Vendida', ascending=True)

# 2. Exibe a lista dos menos vendidos
print("\n--- Produtos Menos Vendidos (Geral) ---")
print(menos_vendidos)

# 3. (Opcional) Salva em um arquivo Excel
menos_vendidos.to_excel("produtos_menos_vendidos.xlsx", index=False)
```

# 5. Frete Médio por Loja

```python
# 1. Calcula a média da coluna “Frete” de cada loja
frete_medio = {
    'Loja 1': loja1['Frete'].mean(),
    'Loja 2': loja2['Frete'].mean(),
    'Loja 3': loja3['Frete'].mean(),
    'Loja 4': loja4['Frete'].mean()
}

# 2. Converte em DataFrame para visualização e exportação
frete_medio_df = (
    pd.Series(frete_medio)
      .reset_index()
      .rename(columns={'index': 'Loja', 0: 'Frete Médio (R$)'})
      .round(2)                       # 2 casas decimais
)

# 3. Salva em uma planilha Excel
frete_medio_df.to_excel('frete_medio_por_loja.xlsx', index=False)

# 4. Exibe o resultado no console
print(frete_medio_df)
```

#Grafico do faturamento

```python
# Faturamento total por loja
fat_tot = pd.Series({
    'Loja 1': loja1['Preço'].sum(),
    'Loja 2': loja2['Preço'].sum(),
    'Loja 3': loja3['Preço'].sum(),
    'Loja 4': loja4['Preço'].sum()
})

# Barra vertical
ax = fat_tot.plot(kind='bar', figsize=(6,4), color='steelblue', title='Faturamento Total por Loja')
ax.set_ylabel('Faturamento (R$)')
ax.bar_label(ax.containers[0], fmt='%.0f')
```

#Participação das categorias (todas as lojas juntas)

```python
# Contagem de vendas por categoria em todas as lojas
cat_share = (
    todas_as_lojas
      .groupby('Categoria do Produto')['Produto']
      .size()
      .sort_values(ascending=False)
)

# Pizza
cat_share.plot(kind='pie',
               autopct='%1.1f%%',
               figsize=(6,6),
               startangle=90,
               ylabel='',
               title='Participação das Categorias (Qtde. de Vendas)')
```

#Preço × Frete (todas as lojas)

```python
import matplotlib.pyplot as plt

# 1. Constrói a tabela “categorias × lojas”
#    (cada célula = faturamento da categoria naquela loja)
#    (ESSA PARTE É IGUAL À ANTERIOR)
tabela_faturamento = (
    pd.concat({
        'Loja1': fat_loja1.set_index('Categoria do Produto')
                          ['Faturamento com o Produto (R$)'],
        'Loja2': fat_loja2.set_index('Categoria do Produto')
                          ['Faturamento com o Produto (R$)'],
        'Loja3': fat_loja3.set_index('Categoria do Produto')
                          ['Faturamento com o Produto (R$)'],
        'Loja4': fat_loja4.set_index('Categoria do Produto')
                          ['Faturamento com o Produto (R$)']
    }, axis=1)
      .fillna(0)
)

# 2. Desenha o gráfico (sem converter para porcentagem)
ax = (tabela_faturamento.T               # linhas = lojas, colunas = categorias
        .plot(kind='bar',
              stacked=True,
              figsize=(9,6),
              colormap='tab20'))         # esquema de cores agradável

# 3. Detalhes visuais
ax.set_title('Faturamento de cada categoria por loja')
ax.set_ylabel('Faturamento (R$)')
ax.set_xlabel('Loja')
plt.legend(title='Categoria', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
